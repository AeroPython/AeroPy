# Result Drawer for the Xfoil Genetic Algorithm

Here are described some useful functions that can help you to graphically display your results.

Place this notebook in the same folder that the Xfoil and the genetic algorithm files.


Here is a list of the included functions:

**profile_read_aero** (generation, profile_number)

Searches for the file of a certain profile generated by the algorithm and return the Cd and Cl.

**profile_read_aero_generic** (root)

Reads a file located in root (ej: calculations/naca6715.txt) and return the Cd and Cl.

**profile_read_alpha** (generation, profile_number)

Searches for the file of a certain profile generated by the algorithm and return the alpha angle and Cl.

**profile_read_alpha_generic** (root)

Reads a file located in root (ej: calculations/naca6715.txt) and return the alpha angle and Cl.

**drawing**(generation, profile_number) 

Searches for the file of a certain profile generated by the algorithm and draws it with matplotlib.

**drawing_bezier**(generation, profile_number)

Searches for the file of a certain profile generated by the algorithm and draws it, along with its bezier points with matplotlib. Useful for understanding the genome of a certain profile.

**drawing_polar**(generation, profile_number)

Plots the Cd against the Cl for a certain profile generated by the algorithm

**drawing_polar_compare**(generation_imput, profile_number_imput)

Plots the Cd against the Cl for some profiles generated by the algorithm. Both 'generation_imput' and 'profile_number_imput' must be numpy arrays of the same dimension.

**drawing_alpha_compare**(generation_imput, profile_number_imput)

Plots the Cl against the alpha for some profiles generated by the algorithm. Both 'generation_imput' and 'profile_number_imput' must be numpy arrays of the same dimension.

**drawing_polar_compare_generic**(profileroots) and **drawing_alpha_compare_generic**(root)

Do the same as the equivalent non-generic functions, but must be fed with an array of file directions (ej: calculations/naca6715.txt)

```python
def profile_read_aero (generation, profile_number):    
    profile_name = 'gen' + str(generation) + 'prof' + str(profile_number)
    data_root = "aerodata\data" + profile_name + '.txt'
    datos = np.loadtxt(data_root, skiprows=12, usecols=[1,2])
    
    read_dim = np.array(datos.shape)
    if ((read_dim.shape[0]) != 2):
        return np.array ([0,0])
    else:
        return datos
```

```python
def profile_read_aero_generic (root):    
    profile_name = root
    data_root = root
    datos = np.loadtxt(data_root, skiprows=12, usecols=[1,2])
    
    read_dim = np.array(datos.shape)
    if ((read_dim.shape[0]) != 2):
        return np.array ([0,0])
    else:
        return datos
```

```python
def profile_read_alpha (generation, profile_number):    
    profile_name = 'gen' + str(generation) + 'prof' + str(profile_number)
    data_root = "aerodata\data" + profile_name + '.txt'
    datos = np.loadtxt(data_root, skiprows=12, usecols=[0,1])
    
    read_dim = np.array(datos.shape)
    if ((read_dim.shape[0]) != 2):
        return np.array ([0,0])
    else:
        return datos
```

```python
def profile_read_alpha_generic (root):    
    
    
    datos = np.loadtxt(root, skiprows=12, usecols=[0,1])
    
    read_dim = np.array(datos.shape)
    if ((read_dim.shape[0]) != 2):
        return np.array ([0,0])
    else:
        return datos
```

```python
def drawing(generation, profile_number): 
    
    profile_root = 'profiles\gen' + str(generation) + '\profile' + str(profile_number) + '.txt'
    
    profile_name = 'Generación ' + str(generation) + ', perfil ' + str(profile_number)
 
    datos = np.loadtxt(profile_root, skiprows=3, usecols=[0,1])
    
    plt.figure(num=None, figsize=(15, 5), dpi=80, facecolor='w', edgecolor='k')
    plt.title (profile_name)
    plt.ylim(-0.15, 0.15)
    plt.xlim(-0.05, 1.05)
    plt.plot(datos[:,0], datos[:,1])
    plt.gca().set_aspect(1)
    
    
    if not os.path.exists('graficos'):
        os.makedirs('graficos')
        
        
    nombre_grafico = 'graficos\gen' + str(generation) + 'profile' + str(profile_number) + '.png'
    plt.savefig(nombre_grafico)
```

```python
def drawing_bezier(generation, profile_number): 
    
    profile_root = 'profiles\gen' + str(generation) + '\profile' + str(profile_number) + '.txt'
    
    profile_name = 'Generación ' + str(generation) + ', perfil ' + str(profile_number)
    
    genome_root = 'genome\generation'+ str(generation) + '.txt'
    
    genome_matrix = np.loadtxt(genome_root, skiprows=1)
    
    genome = genome_matrix[profile_number-1,:]
    
    bezier_points = generador_puntos(genome)
 
    datos = np.loadtxt(profile_root, skiprows=3, usecols=[0,1])
    
    plt.figure(num=None, figsize=(15, 5), dpi=80, facecolor='w', edgecolor='k')
    plt.title (profile_name)
    plt.ylim(-0.15, 0.15)
    plt.xlim(-0.05, 1.05)
    plt.scatter(bezier_points[:,0],bezier_points[:,1])
    plt.plot(datos[:,0], datos[:,1])
    plt.plot(bezier_points[0:2 , 0] , bezier_points[0:2 , 1])
    plt.plot(bezier_points[2:5 , 0] , bezier_points[2:5 , 1])
    plt.plot(bezier_points[5:8 , 0] , bezier_points[5:8 , 1])
    plt.plot(bezier_points[8:11 , 0] , bezier_points[8:11 , 1])
    plt.plot(bezier_points[11:13 , 0] , bezier_points[11:13 , 1])
    plt.gca().set_aspect(1)
    
    
    if not os.path.exists('graficos'):
        os.makedirs('graficos')
        
        
    nombre_grafico = 'graficos\gen' + str(generation) + 'profile' + str(profile_number) + 'bezierpoints.png'
    plt.savefig(nombre_grafico)
```

```python
def drawing_polar(generation, profile_number): 
    
    profile_root = 'profiles\gen' + str(generation) + '\profile' + str(profile_number) + '.txt'
    
    profile_name = 'Generación ' + str(generation) + ', perfil ' + str(profile_number)
 
    datos = profile_read_aero(generation, profile_number)
    
    plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
    plt.title (profile_name)
    plt.plot(datos[:,0], datos[:,1])
    
    
    if not os.path.exists('graficos'):
        os.makedirs('graficos')
        
        
    nombre_grafico = 'graficos\gen' + str(generation) + 'profile' + str(profile_number) + 'polar.png'
    plt.savefig(nombre_grafico)
```

```python
def drawing_polar_compare(generation_imput, profile_number_imput): 
    
    num_prof = generation_imput.shape[0]
    profile_name = 'Generación ' + str(generation_imput) + ', perfil ' + str(profile_number_imput)
    plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
    plt.title (profile_name)
    
    for i in np.arange(0,num_prof,1):
        generation = generation_imput[i]
        profile_number = profile_number_imput[i]
    
        profile_root = 'profiles\gen' + str(generation) + '\profile' + str(profile_number) + '.txt'
    
        
 
        datos = profile_read_aero(generation, profile_number)
    
       
       
        plt.plot(datos[:,0], datos[:,1])
    
    
    if not os.path.exists('graficos'):
        os.makedirs('graficos')
        
        
    nombre_grafico = 'graficos\gen' + str(generation_imput) + 'profile' + str(profile_number_imput) + 'polar.png'
    plt.savefig(nombre_grafico)
```

```python
def drawing_alpha_compare(generation_imput, profile_number_imput): 
    
    num_prof = generation_imput.shape[0]
    profile_name = 'Generación ' + str(generation_imput) + ', perfil ' + str(profile_number_imput)
    plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
    plt.title (profile_name)
    
    ideal = np.array([[0,0],
                      [15, (15*np.pi/180)*np.pi*2]])
        
    plt.plot(ideal[:,0], ideal[:,1])
    for i in np.arange(0,num_prof,1):
        generation = generation_imput[i]
        profile_number = profile_number_imput[i]
    
        profile_root = 'profiles\gen' + str(generation) + '\profile' + str(profile_number) + '.txt'
    
        
 
        datos = profile_read_alpha(generation, profile_number)
    
       
       
        plt.plot(datos[:,0], datos[:,1])
    
    
    if not os.path.exists('graficos'):
        os.makedirs('graficos')
        
        
    nombre_grafico = 'graficos\gen' + str(generation_imput) + 'profile' + str(profile_number_imput) + 'alpha.png'
    plt.savefig(nombre_grafico)
```

```python
def drawing_polar_compare_generic(profileroots): 
    
    num_prof = profileroots.shape[0]
    
    plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
    plt.rc('font', size = 20)
    plt.title ('Profile comparison')
    
    for i in np.arange(0,num_prof,1):
        
    
        profile_root = profileroots[i]
    
        
 
        datos = profile_read_aero_generic(profile_root)
    
       
       
        plt.plot(datos[:,0], datos[:,1], label = profile_root)
    
    
    if not os.path.exists('graficos'):
        os.makedirs('graficos')
        
    plt.legend(loc = 2, fontsize =14)  
    plt.grid()
    plt.minorticks_on()
    plt.xlabel('Lift coefficient')  
    plt.ylabel('Drag coefficient')
    nombre_grafico = 'graficos\comparepolar.png'
    plt.savefig(nombre_grafico)
```

```python
def drawing_alpha_compare_generic(root): 
    
    num_prof = root.shape[0]
    
    plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
    plt.rc('font', size = 20)
    plt.title ('profile comparison')
    
    ideal = np.array([[0,0],
                      [15, (15*np.pi/180)*np.pi*2]])
        
    plt.plot(ideal[:,0], ideal[:,1], label = ' ideal, cl alpha = 2$\pi$')
    for i in np.arange(0,num_prof,1):
       
        profile_root = root[i]
    
        
 
        datos = profile_read_alpha_generic(profile_root)
    
       
       
        plt.plot(datos[:,0], datos[:,1], label = profile_root)
    
    
    if not os.path.exists('graficos'):
        os.makedirs('graficos')
        
    plt.legend(loc = 2, fontsize =14)  
    plt.grid() 
    plt.minorticks_on()
    plt.xlabel('Attack angle')  
    plt.ylabel('Lift coefficient')
    nombre_grafico = 'graficos\compare_alpha.png'
    plt.savefig(nombre_grafico)
```
